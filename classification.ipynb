```python
import zipfile
import pandas as pd
import numpy as np
```

```python
with zipfile.ZipFile('data.zip', 'r') as zip_ref:
    zip_ref.extractall('.')

data = pd.read_csv('smoke_detection_iot.csv')
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0', 'CNT'])
```

```python
data['Fire Alarm'].value_counts()
```

```python
data.describe()
```

```python
df_0 = data[data['Fire Alarm'] == 0].head(10000)

df_1 = data[data['Fire Alarm'] == 1].head(10000)

data = pd.concat([df_0, df_1])

data.head()
```

```python
data['Fire Alarm'].value_counts()
```

```python
data.describe()
```

```python
X = data[data.columns[:-1]]
y = data['Fire Alarm']
X
```

```python
# Scale
from sklearn.preprocessing import StandardScaler

fields = X.columns
scaler = StandardScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X, columns=fields)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train
```

### KNN for Classification
Using Grid Search - Cross Validation to determine the best hyperparameters

```python
# Hyperparameters 1

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Trying different values for K
    'weights': ['uniform', 'distance'],  # Uniform weighting vs distance weighting
    'metric': ['euclidean', 'manhattan']  # Trying different distance metrics
}

knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)  # 5-fold cross-validation
grid_search.fit(X_train, y_train)
y_pred_knn = grid_search.predict(X_test)

print(f"Best parameters: {grid_search.best_params_}")
```

```python
y_pred_knn = pd.DataFrame(y_pred_knn, columns=['y_pred_knn'])
results_knn = pd.concat([X_test.reset_index(drop=True), y_pred_knn], axis=1)
results_knn.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))
```

# Decision Tree for Classification


```python
# # Decision Tree for Classification
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
# y_pred_tree = dt.predict(X_test)

dt.tree_.node_count, dt.tree_.max_depth
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def measure_error(y_true, y_pred, label):
    return pd.Series({'accuracy': accuracy_score(y_true, y_pred),
                      'precision_score': precision_score(y_true, y_pred, average='macro'),
                      'recall_score': recall_score(y_true, y_pred,average='macro'),
                      'f1_score': f1_score(y_true, y_pred,average='macro')},
                      name=label)
```

Tried first with no limits on maximum depth, features and leaves

```python
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

train_test_full_error = pd.concat([measure_error(y_train, y_train_pred, 'train'),
                                   measure_error(y_test, y_test_pred, 'test')],
                                   axis=1)
train_test_full_error
```

Then, try with Grid Search fit hyperparameters

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth':range(1, dt.tree_.max_depth+1, 2),
              'max_features':range(1, len(dt.feature_importances_)+1)}

GR = GridSearchCV(DecisionTreeClassifier(random_state=42),
                  param_grid=param_grid,
                  scoring='accuracy',
                  n_jobs=-1)

GR = GR.fit(X_train, y_train)
```

```python
GR.best_estimator_.tree_.node_count, GR.best_estimator_.tree_.max_depth
```

```python
y_train_pred_gr = GR.predict(X_train)
y_test_pred_gr = GR.predict(X_test)

train_test_gr_error = pd.concat([measure_error(y_train, y_train_pred_gr, 'train'),
                                   measure_error(y_test, y_test_pred_gr, 'test')],
                                   axis=1)
train_test_gr_error
```

## Metrics
Difference between two methods

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_test, y_test_pred_gr))
print(classification_report(y_test, y_test_pred_gr))
print(confusion_matrix(y_test, y_test_pred_gr))
```
